```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
hh_tiff=("D:\\Digital Image Processing\\ALPSRP243410540-H2.2_UA\\HH-ALPSRP243410540-H2.2_UA.tif")
hv_tiff=("D:\\Digital Image Processing\\ALPSRP243410540-H2.2_UA\\HV-ALPSRP243410540-H2.2_UA.tif")
with rasterio.open(hh_tiff) as hh_dataset, rasterio.open(hv_tiff) as hv_dataset:
    hh_data= hh_dataset.read(1)
    hv_data= hv_dataset.read(1)
hh_norm=(hh_data-np.min(hh_data))/(np.max(hh_data) - np.min(hh_data))
hv_norm=(hv_data-np.min(hv_data))/(np.max(hv_data) - np.min(hv_data))
diff=(hh_norm-hv_norm)
colormap=plt.get_cmap("seismic")
diff_colormap=(colormap(diff)[:,:,:3]*255).astype(np.int16)
plt.figure(figsize=(10,10))
plt.imshow(diff_colormap)
plt.title("HH-HV Difference Composite")
plt.axis('ON')
plt.show()
```
